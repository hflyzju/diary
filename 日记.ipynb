#### 一：K均值聚类
>- 1.首先选择k歌点作为起始质心，然后对数据集中所有的点算出离这k个点的距离，并把他归为离他最近的那一类去，这一步完成后，每个簇的质心更新为该簇点所有点的平均值，直到所有点的簇分布结果不发生变化时停止循环。
>- 2.因为第一步找出来的簇心可能是局部最小值，而非全局最小值（如机器学习实战190页图），所以出现了一种度量聚类效果的指标即SSE（Sum of Squared Error误差平方和）。但是由于如果增加k的数量，SSE的值肯定会降低失去判断价值，所以聚类的目标是在保持簇数不变得情况下提高簇的质量
>- 3.二分K-均值算法，该算法首先将所有点作为一个簇，然后将该簇一分为二，之后选择其中一个簇继续进行划分，选择哪一个簇进行划分取决于对其划分是否可以最大程度降低SSE的值。上述基于SSE的划分过程不断重复，知道得到用户指定的簇数目为止。

### 二：关联分析之Apriori算法

>- 1.**支持度：**数据集中包含该相集的记录所占的比例。**置信度：**支持度{尿布，葡萄酒}/支持度{尿布}=0.75，意味着对于包含尿布的所有记录，我们的规则对其中75%的记录都适用
>- 2.Apriori算法示例如下（摘自王灿老师PPT），最小支持度为50%：![Apriori](http://zhangxiaoyang.me/drafts/categories/data-mining-tutorial/images/apriori.jpg)
>- 3.Apriori算法的主要开销在于产生大量候选集和重复扫描数据库,改进之FP-tree算法如下：![FP-tree](http://zhangxiaoyang.me/drafts/categories/data-mining-tutorial/images/FP-tree.jpg)![FP-tree](http://zhangxiaoyang.me/drafts/categories/data-mining-tutorial/images/%E7%BB%88%E6%9E%81FP-tree.jpg)
>- 4.挖掘是从项头表的底部开始，即支持度计数小的开始，也就是从p开始。我们可以得到p-条件模式基：fcam:2、cb:1。可以看到，挖掘的方法就是从根部到p的路径上，记录下沿途经过的节点。对于p来说，有两条路径到达p，因为有2个p。所以就有了fcam和cb，2和1是由p决定的。通过这样的方式，就可以分别得到p、m、b、a、c、f的条件模式基。

### 三：降维之PCA

>- 1.线性代数里面求特征矩阵来降维,总的来说是使高维的坐标映射到地维上去，并使其各个点分开的距离最大，及不相关性越大越好
>- 2.步骤：

>- 1）将原始数据按列组成n行m列矩阵X
>- 2）将X的每一行（代表一个属性字段）进行零均值化，即减去这一行的均值
>- 3）求出协方差矩阵C=1mXXTC=1mXXT
>- 4）求出协方差矩阵的特征值及对应的特征向量
>- 5）将特征向量按对应特征值大小从上到下按行排列成矩阵，取前k行组成矩阵P
>- 6）Y=PXY=PX即为降维到k维后的数据

>- 3.可以参考后面的实例：[PCA的数学原理]：http://blog.codinglabs.org/articles/pca-tutorial.html

### 光流法总结
>- 1.博客地址：http://www.cnblogs.com/scut-fm/p/3748007.html、
>- 2.运动物体的灰度在很短的间隔时间内保持不变，从而得到一个有x，y方向上俩个变量的方程。要求解，还需要一个方程。
>- 3.假设给定邻域内的速度向量场变化是缓慢的，得到一个速度在x，y方向上的二阶导数最小化，然后再根据欧拉方程求解。
>- 4.对于火焰，乘了一个光强强度I，最后再和上述方法一样求解。
>- 5.最后还是结合了经典的光流法来得出几个特征向量，来判断火焰。
